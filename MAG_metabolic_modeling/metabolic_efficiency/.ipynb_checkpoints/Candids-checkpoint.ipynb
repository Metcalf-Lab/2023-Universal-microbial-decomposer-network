```python
import pandas as pd
```

```python
Main=pd.read_excel('Media.xlsx',sheet_name='uptake')
Main
```

```python
from collections import Counter
Complete=Main.iloc[1:,[0,3]]
Data=Main.iloc[1:,[0,3,4]]
```

```python
Data.iloc[59,:]
```

```python
pd.set_option('display.max_rows', 93)
Data=Data.rename(columns={'community metabolites':'CM','Unnamed: 3':'MW','Unnamed: 4':'Type'})
Data.reset_index(drop='index')
```

```python
AAs=Data.loc[0:20]
AAs=AAs.to_dict()
Carbs=Data.loc[21:32]
Carbs=Carbs.to_dict()
Fiber=Data.loc[33:44]
Fiber=Fiber.to_dict()
Lipid=Data.loc[45:54]
Lipid=Lipid.to_dict()
Minerals=Data.loc[56:73]
Minerals=Minerals.to_dict()
Vitamins=Data.loc[74:91]
Vitamins=Vitamins.to_dict()
PolyAmine=Data.loc[92]
PolyAmine=PolyAmine.to_dict()
Water=Data.loc[93]
Water=Water.to_dict()

```

```python
import json as json

with open('AAs.json', 'w') as json_file:
  json.dump(AAs, json_file)

with open('Carbs.json', 'w') as json_file:
  json.dump(Carbs, json_file)

with open('Fiber.json', 'w') as json_file:
  json.dump(Fiber, json_file)

with open('Lipid.json', 'w') as json_file:
  json.dump(Lipid, json_file)

with open('Minerals.json', 'w') as json_file:
  json.dump(Minerals, json_file)

with open('Vitamins.json', 'w') as json_file:
  json.dump(Vitamins, json_file)

with open('PolyAmine', 'w') as json_file:
  json.dump(PolyAmine, json_file)

with open('Water.json', 'w') as json_file:
  json.dump(Water, json_file)
```
