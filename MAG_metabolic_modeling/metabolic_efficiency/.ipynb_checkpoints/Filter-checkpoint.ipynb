```python
import pandas as pd
```

```python
OTU_Table=pd.read_csv("shotgun_soil_gOTU_table.txt",sep='\t')
OTU_Table
```

```python
Rel_Abund=OTU_Table
for i in range(len(OTU_Table.columns)-1):
    Rel_Abund.iloc[0:,i+1]=OTU_Table.iloc[0:,i+1]/OTU_Table.iloc[0:,i+1].sum()
for i in range(len(OTU_Table.columns)-1):
    Rel_Abund.columns=range(len( Rel_Abund.columns))
Rel_Abund
```

```python
K=50
Cols=list(Rel_Abund.columns)[1:]
Candidates={}
Candidates_Rel={}
for Columns in Cols:
    Filt=Rel_Abund[Columns].sort_values(ascending=False)>0
    if len(Rel_Abund.loc[Filt])>K:
        Temp=Rel_Abund.loc[Filt,Columns].sort_values(ascending=False)
        Candidates[Columns]=list(Rel_Abund.iloc[Temp.head(K).index,0])
    else:
        Candidates[Columns]=list(Rel_Abund.loc[Filt,"#OTU ID"])
    
   
    
Candidates
# for columns in Rel_Abund.columns 
```

```python
import json as json

with open('Candidates.json', 'w') as json_file:
  json.dump(Candidates, json_file)
```
