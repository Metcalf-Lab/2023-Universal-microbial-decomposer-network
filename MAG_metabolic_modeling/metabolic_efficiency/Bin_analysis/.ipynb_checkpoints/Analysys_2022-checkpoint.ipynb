```python
import pandas as pd
import seaborn as sns
import re
from sklearn import linear_model
from matplotlib import pyplot as plt
import numpy as np
import statsmodels.api as sm
from scipy import stats
import xlsxwriter
```

```python
Main_Table=pd.read_table('Bins_ATP.csv',delimiter=',',index_col='Row')
Main_Table["Site"]='Null'
Main_Table=Main_Table[Main_Table[['Carbohydrates','AminoAcids','Lipids']]<1]
Main_Table.loc[Main_Table.index.str.contains('UTK'),["Site"]]="UTK"
Main_Table.loc[Main_Table.index.str.contains('SHSU'),["Site"]]="SHSU"
Main_Table.loc[Main_Table.index.str.contains('CMU'),["Site"]]="CMU"
Time_Loc_Table=pd.read_table('Jan2022_mapping/MAG_table_final_models.tsv',delimiter='\t',header=1,index_col="#OTU ID")/1000000
Meta=pd.read_table('Jan2022_mapping/shotgun_metadata_r.txt',delimiter='\t',header=0,index_col='SampleID')
Main_Table

Meta.to_csv("Jan2022_mapping/Metadata.csv")
Samples=list(Time_Loc_Table.columns[1:])
```

```python
# for sample in Samples

Main_Table.set_axis([ind[:-4] for ind in list(Main_Table.index)],inplace=True)
Main_Table.drop(["Site"],axis=1,inplace=True)
Main_Table.fillna(0,inplace=True)
```

```python
Bins=list(Main_Table.index)
Time_Loc_Table
```

```python
Main_Table
```

```python
Contribution=pd.DataFrame()
counter=0
writer = pd.ExcelWriter('Jan2022_mapping/Contribution.xlsx', engine='xlsxwriter')
for sample in list(Time_Loc_Table.columns):
    for Bin in list(Main_Table.index):
        if list(Time_Loc_Table.index).__contains__(Bin):
            Contribution.loc[Bin,['Carbohydrates','Lipids','AminoAcids']]=list(Main_Table.loc[Bin,['Carbohydrates','Lipids','AminoAcids']].multiply(Time_Loc_Table.loc[Bin,sample]))
            counter+=1
    Contribution.to_excel(writer, sheet_name=sample[0:30])
writer.save()
```

```python
Contribution
```

```python
ATP={}
for sample in Samples:
    ATP[sample]=[0,0,0]
    for Bin in Bins:
        if list(Time_Loc_Table.index).__contains__(Bin):
            if Time_Loc_Table.loc[Bin,sample]>0:
                ATP[sample]+=Main_Table.loc[Bin,:]*Time_Loc_Table.loc[Bin,sample]
df_ATP=pd.DataFrame.from_dict(ATP)
df_ATP=df_ATP.T
df_ATP[["facility","State"]]="Empty"
for inds in df_ATP.index:
    df_ATP.loc[inds,'facility']=Meta.loc[inds,'facility']
    df_ATP.loc[inds,'State']=Meta.loc[inds,'add_0c_group']
    df_ATP.loc[inds,'Conv_time']=Meta.loc[inds,'add_0c']
```

```python
ax = sns.boxplot(x="facility", y="Lipids",data=df_ATP)

ax = sns.swarmplot(x="facility", y="Lipids",  color=".25", data=df_ATP)
plt.savefig("Jan2022_mapping/BoxandSwarmplots.png",dpi=600)
plt.show()
```

```python
Media=['Lipids','Carbohydrates','AminoAcids']
Facility=["CMU","SHSU","UTK"]
color=["purple","blue","green","purple","blue","green","purple","blue","green"]
count=0
np.set_printoptions(precision=3)
fig, axs = plt.subplots(3, 3)
for i in Media:
    for j in Facility:                              
        df=df_ATP[df_ATP["facility"]==j]  
        sns.regplot(x="Conv_time", y=i, data=df,ax=axs[Media.index(i), Facility.index(j)],scatter_kws={'s':3}, color=color[0+count])
        x_pos=axs[Media.index(i), Facility.index(j)].get_xlim()[1]-axs[Media.index(i), Facility.index(j)].get_xlim()[1]/20
        y_pos=axs[Media.index(i), Facility.index(j)].get_ylim()[1]-axs[Media.index(i), Facility.index(j)].get_ylim()[1]/20
        exec('axs[Media.index(i), Facility.index(j)].set_title('+"f'{i} Efficiency ({j})')")
        pearson_coef, p_value = stats.pearsonr(df["Conv_time"], df[i])
        T='P_value: {}\nCorrelation Coef.: {}'.format(np.format_float_scientific(p_value,3),np.format_float_positional(pearson_coef,3))
        axs[Media.index(i), Facility.index(j)].text(5,0.69,T,verticalalignment='top', horizontalalignment='left')
        count+=1
        
for ax in axs.flat:
    ax.set_ylabel("ATP per C-mol Substrate")
    ax.set_xlabel("ADD")
    ax.set_xlim(0,575)
    ax.set_ylim(0,0.7)

# fig.tight_layout()
fig.set_figheight(15)
fig.set_figwidth(15)
plt.savefig("Jan2022_mapping/Subplots_With_Correlation_line.png",dpi=600,bbox_inches='tight')
```
