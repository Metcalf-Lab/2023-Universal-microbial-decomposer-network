```python
import pandas as pd
import seaborn as sns
import re
from sklearn import linear_model
from matplotlib import pyplot as plt
import numpy as np
```

```python
Main_Table=pd.read_table('Bins_ATP.csv',delimiter=',',index_col='Row')
Main_Table["Site"]='Null'
Main_Table=Main_Table[Main_Table[['Carbs','AAs','Lipids']]<1]
Main_Table.loc[Main_Table.index.str.contains('UTK'),["Site"]]="UTK"
Main_Table.loc[Main_Table.index.str.contains('SHSU'),["Site"]]="SHSU"
Main_Table.loc[Main_Table.index.str.contains('CMU'),["Site"]]="CMU"
Time_Loc_Table=pd.read_table('zach_genome_tpms_final.tsv',delimiter='\t',header=1,index_col="#OTU ID")/1000000
Meta=pd.read_table('shotgun_metadata_condensed.txt',delimiter='\t',header=0,index_col='SampleID')
Main_Table

Meta.to_csv("Metadata.csv")
Samples=list(Time_Loc_Table.columns[1:])
```

```python
# for sample in Samples

Main_Table.set_axis([ind[:-4] for ind in list(Main_Table.index)],inplace=True)
Main_Table.drop(["Site"],axis=1,inplace=True)
Main_Table.fillna(0,inplace=True)
```

```python
Bins=list(Main_Table.index)
Time_Loc_Table
```

```python
ATP={}
for sample in Samples:
    ATP[sample]=[0,0,0]
    for Bin in Bins:
        if list(Time_Loc_Table.index).__contains__(Bin):
            if Time_Loc_Table.loc[Bin,sample]>0:
                ATP[sample]+=Main_Table.loc[Bin,:]*Time_Loc_Table.loc[Bin,sample]
df_ATP=pd.DataFrame.from_dict(ATP)
df_ATP=df_ATP.T
df_ATP[["facility","State"]]="Empty"
for inds in df_ATP.index:
    df_ATP.loc[inds,'facility']=Meta.loc[inds,'facility']
    df_ATP.loc[inds,'State']=Meta.loc[inds,'add_0c_group']
    df_ATP.loc[inds,'Conv_time']=Meta.loc[inds,'add_0c']
```

```python
ax = sns.boxplot(x="facility", y="Lipids",data=df_ATP)

ax = sns.swarmplot(x="facility", y="Lipids",  color=".25", data=df_ATP)
```

```python
CMU_Filt=df_ATP["facility"]=="CMU"
SHSU_FIlt=df_ATP["facility"]=="SHSU"
UTK_FIlt=df_ATP["facility"]=="UTK"
Lipid_CMU=list(df_ATP["Lipids"].loc[CMU_Filt])
Lipid_CMU_Day=list(df_ATP["Conv_time"].loc[CMU_Filt])
Lipid_SHSU=list(df_ATP["Lipids"].loc[SHSU_FIlt])
Lipid_SHSU_Day=list(df_ATP["Conv_time"].loc[SHSU_FIlt])
Lipid_UTK=list(df_ATP["Lipids"].loc[UTK_FIlt])
Lipid_UTK_Day=list(df_ATP["Conv_time"].loc[UTK_FIlt])
Carbohydrates_CMU=list(df_ATP["Carbs"].loc[CMU_Filt])
Carbohydrates_CMU_Day=list(df_ATP["Conv_time"].loc[CMU_Filt])
Carbohydrates_SHSU=list(df_ATP["Carbs"].loc[SHSU_FIlt])
Carbohydrates_SHSU_Day=list(df_ATP["Conv_time"].loc[SHSU_FIlt])
Carbohydrates_UTK=list(df_ATP["Carbs"].loc[UTK_FIlt])
Carbohydrates_UTK_Day=list(df_ATP["Conv_time"].loc[UTK_FIlt])
AminoAcids_CMU=list(df_ATP["AAs"].loc[CMU_Filt])
AminoAcids_CMU_Day=list(df_ATP["Conv_time"].loc[CMU_Filt])
AminoAcids_SHSU=list(df_ATP["AAs"].loc[SHSU_FIlt])
AminoAcids_SHSU_Day=list(df_ATP["Conv_time"].loc[SHSU_FIlt])
AminoAcids_UTK=list(df_ATP["AAs"].loc[UTK_FIlt])
AminoAcids_UTK_Day=list(df_ATP["Conv_time"].loc[UTK_FIlt])

fig, axs = plt.subplots(3, 3)
axs[0, 0].scatter(Lipid_CMU_Day,Lipid_CMU,s=5 )
axs[0, 0].set_title('Lipid_CMU')
axs[1, 0].scatter(Lipid_SHSU_Day,Lipid_SHSU,s=5 )
axs[1, 0].set_title('Lipid_SHSU')
axs[2, 0].scatter(Lipid_UTK_Day,Lipid_UTK,s=5 )
axs[2, 0].set_title('Lipid_UTK')
axs[0, 1].scatter(Carbohydrates_CMU_Day,Carbohydrates_CMU,s=5 )
axs[0, 1].set_title('Carbohydrates_CMU')
axs[1, 1].scatter(Carbohydrates_SHSU_Day,Carbohydrates_SHSU,s=5 )
axs[1, 1].set_title('Carbohydrates_SHSU')
axs[2, 1].scatter(Carbohydrates_UTK_Day,Carbohydrates_UTK,s=5 )
axs[2, 1].set_title('Carbohydrates_UTK')
axs[0, 2].scatter(AminoAcids_CMU_Day,AminoAcids_CMU,s=5 )
axs[0, 2].set_title('AminoAcids_CMU')
axs[1, 2].scatter(AminoAcids_SHSU_Day,AminoAcids_SHSU,s=5 )
axs[1, 2].set_title('AminoAcids_SHSU')
axs[2, 2].scatter(AminoAcids_UTK_Day,AminoAcids_UTK,s=5 )
axs[2, 2].set_title('AminoAcids_UTK')

for ax in axs.flat:
    ax.set_ylabel("ATPm")
    ax.set_xlabel("Conv_time")
    
# fig.tight_layout()
fig.set_figheight(17)
fig.set_figwidth(15)
plt.savefig("Subplots.png",dpi=1200)
```

```python
regr = linear_model.LinearRegression()
AA_UTK_Day=np.array(AminoAcids_UTK_Day)
regr.fit(AA_UTK_Day[:,np.newaxis],AminoAcids_UTK)
plt.scatter(AminoAcids_UTK_Day,AminoAcids_UTK,s=5 )
plt.plot(AminoAcids_UTK_Day,regr.predict(AA_UTK_Day[:,np.newaxis]))
```
