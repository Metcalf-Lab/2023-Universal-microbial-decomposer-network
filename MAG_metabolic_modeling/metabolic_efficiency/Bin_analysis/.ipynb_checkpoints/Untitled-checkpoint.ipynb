```python
import pandas as pd
import seaborn as sns
import re

from matplotlib import pyplot as plt
```

```python
Main_Table=pd.read_table('Bins_ATP.csv',delimiter=',',index_col='Row')
Main_Table["Site"]='Null'
Main_Table=Main_Table[Main_Table[['Carbs','AAs','Lipids']]<1]
Main_Table.loc[Main_Table.index.str.contains('UTK'),["Site"]]="UTK"
Main_Table.loc[Main_Table.index.str.contains('SHSU'),["Site"]]="SHSU"
Main_Table.loc[Main_Table.index.str.contains('CMU'),["Site"]]="CMU"
Time_Loc_Table=pd.read_table('zach_genome_tpms_final.tsv',delimiter='\t',header=1,index_col="#OTU ID")
Meta=pd.read_table('shotgun_metadata_condensed.txt',delimiter='\t',header=0,index_col='SampleID')
Main_Table

Meta.to_csv("Metadata.csv")
Samples=list(Time_Loc_Table.columns[1:])
```

```python
# for sample in Samples

Main_Table.set_axis([ind[:-4] for ind in list(Main_Table.index)],inplace=True)
Main_Table.drop(["Site"],axis=1,inplace=True)
Main_Table.fillna(0,inplace=True)
```

```python
Bins=list(Main_Table.index)
Time_Loc_Table
```

```python
ATP={}
for sample in Samples:
    ATP[sample]=[0,0,0]
    for Bin in Bins:
        if Time_Loc_Table.loc[Bin,sample]>0:
            ATP[sample]+=ATP[sample]+Main_Table.loc[Bin]*Time_Loc_Table.loc[Bin,sample]
ATP
```

```python
ax = sns.boxplot(x="Site", y="AAs",data=Main_Table)

ax = sns.swarmplot(x="Site", y="AAs",  color=".25", data=Main_Table)
```
