## CMU

```python
Metab_winter_CMU = Metab_winter.filter_samples('facility', 'CMU')
```

```python
Metab_winter_CMU.sample_metadata.season.value_counts()
```

```python
Metab_winter_CMU.sample_metadata.facility.value_counts()
```

```python
Metab_winter_CMU.sample_metadata.subjectID.value_counts()
```

```python
gkf = GroupKFold(3)
Metab_winter_CMU = Metab_winter_CMU.add_sample_metadata_as_features(['avg_percent_humidity','total_precipitation_cm'])

X_winter_CMU = Metab_winter_CMU.data
y_winter_CMU = Metab_winter_CMU.sample_metadata['add_0c']
y_winter_CMU = (y_winter_CMU.astype(float))
groups = Metab_winter_CMU.sample_metadata['subjectID']
# Tested this param grid on Summit to expand hyperparameter tuning
#param_grid = {"max_depth": [4, 8, 16, None],
#              "max_features": ['sqrt', 'log2', 0.1],
#              "min_samples_split": [0.001, 0.01, 0.1],
#              "min_weight_fraction_leaf": [0.0001, 0.001, 0.01],
#              "bootstrap": [True, False]}
# It provided this best model:
          
# Takes too long to run on local computer; will need to run then import model from Summit

param_grid = {"max_depth": [4],
          "max_features": ['log2'],
          "min_samples_split": [0.1],
          "min_weight_fraction_leaf": [0.0001],
          "bootstrap": [False]}
rf = RandomForestRegressor(n_estimators=1000, random_state=999, criterion='mae')
gs = GridSearchCV(rf, param_grid=param_grid, cv=gkf.split(X_winter_CMU, y_winter_CMU, groups), scoring='neg_mean_absolute_error', n_jobs=-3)
```

```python
gs.fit(X_winter_CMU, y_winter_CMU)
```

```python
print(gs.best_params_)
```

```python
print('The best error rate is', -gs.best_score_)

with open('soil_model_errors.txt', 'a') as out:
   out.write('winter_CMU_soil\t'+str(round(-gs.best_score_,2))+'\n')
out.close
```

```python
joblib.dump(gs.best_estimator_, 'models/PMI3_Metab_winter_CMU_soil.pkl')
```

```python
model_Metab_winter_CMU = joblib.load('models/PMI3_Metab_winter_CMU_soil.pkl')
model_Metab_winter_CMU
```

## Recursive Feature Elimination 

### Uses the best estimator from the above model and performs 5% removal recursive steps

```python
rfecv = RFECV(estimator=model_Metab_winter_CMU, step=0.05, min_features_to_select=100, cv=list(gkf.split(X_winter_CMU, y_winter_CMU, groups)), scoring='neg_mean_absolute_error', n_jobs=-3)
```

```python
# Likely need to run on Summit
rfecv.fit(X_winter_CMU, y_winter_CMU)
```

```python
print("Number of steps performed: %d" % len(rfecv.grid_scores_))
print("Optimal number of features : %d" % rfecv.n_features_)
print("The best mean absolute error is: %d" % -(max(rfecv.grid_scores_)))
```

```python
# Plot number of features VS. cross-validation scores
plt.figure(figsize=(8,5))
plt.title('Recursive Feature Elimination with Cross-Validation', fontsize=14, fontweight='bold', pad=20)
plt.xlabel("Number of Steps (5% of features added per step)")
plt.ylabel("Cross Validation Score (Mean Absolute Error)")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_, linewidth=2)
plt.savefig('figures/winter_CMU_soil_model_rfecv.png')
```

```python
# Saves the estimator fit on the reduced dataset (i.e. the model created from optimal features)
joblib.dump(rfecv.estimator_, 'models/PMI3_Metab_soil_winter_CMU_rfecv.pkl')
```

```python
model_Metab_winter_CMU_rfecv = joblib.load('models/PMI3_Metab_soil_winter_CMU_rfecv.pkl')
model_Metab_winter_CMU_rfecv
```

### Predicting own data (will be overfit but good sanity check)

```python
# Removes the unimportant features from X
X_winter_CMU = rfecv.transform(X_winter_CMU)

yhat = model_Metab_winter_CMU_rfecv.predict(X_winter_CMU)
MAE = mean_absolute_error(y_winter_CMU, yhat)
print('Mean Absolute Error:', round(MAE, 2), 'ADD')

#Score of best estimator when used to estimate training data (will be overfit)
print('Coefficient of determination R^2 of the prediction:', round(model_Metab_winter_CMU_rfecv.score(X_winter_CMU,y_winter_CMU),2))
```

```python
plt.scatter(yhat, y_winter_CMU, color = 'b')
lower = np.min(np.concatenate([y_winter_CMU, yhat]))
upper = np.max(np.concatenate([y_winter_CMU, yhat]))
plt.plot([lower, upper], [lower, upper], color = 'black', lw = 2, ls = ':')
plt.xlim(0,600)
plt.ylim(0,600)

#regression part
slope, intercept, r_value, p_value, std_err = stats.linregress(yhat,y_winter_CMU)
line = slope*yhat+intercept
plt.plot(yhat, line, color = 'black', lw = 2)

plt.title('Winter CMU soil model pred self\n MAE = %d; $R^2$ = %.2f' %(round(MAE, 2),round(model_Metab_winter_CMU_rfecv.score(X_winter_CMU,y_winter_CMU),2)))
plt.xlabel("Predicted ADD")
plt.ylabel("Actual ADD")
plt.savefig('figures/winter_CMU_soil_model_self.png')
```

### Determine important features after RFECV

```python
importances = model_Metab_winter_CMU_rfecv.feature_importances_
std = np.std([tree.feature_importances_ for tree in model_Metab_winter_CMU_rfecv.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

feature_metadata = Metab_winter_CMU.feature_metadata
np.savetxt("feature_metadata/winter_CMU_soil_feature_metadata_rfecv.csv", feature_metadata, delimiter=",", fmt='%s')


#print the important ids in order
soil_importances = []
count = 0
print("Feature:\t\t\t\tImportance:")
for i in indices:
    soil_importances += (Metab_winter_CMU.feature_metadata.index.values[i], importances[indices[count]])
    if count < 25:
        print(str(count+1)+". "+str(Metab_winter_CMU.feature_metadata.index.values[i])+"\t"+str(importances[indices[count]]))
    count += 1
```

```python
# make sure shape is right
print("Number of features present:", int(len(soil_importances)/2))
soil_importances_df = pd.DataFrame(np.array(soil_importances).reshape(int(len(soil_importances)/2),2))

np.savetxt("importances/winter_CMU_soil_importances.csv", soil_importances_df, delimiter=",", fmt='%s')
```
